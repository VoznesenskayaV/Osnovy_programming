**2. Зависит ли длительность песен от оплаты?**

Исследуем, существует ли зависимость между длительностью песен и их стоимостью, вычислив коэффициент корреляции Пирсона и выведем результат:

```python
import pandas as pd
from scipy.stats import pearsonr

# Загрузка данных из файла Excel
df_artists = pd.read_excel('BD.xlsx', sheet_name='Артисты')
df_albums = pd.read_excel('BD.xlsx', sheet_name='Альбомы')
df_tracks = pd.read_excel('BD.xlsx', sheet_name='Треки')
df_genres = pd.read_excel('BD.xlsx', sheet_name='Жанры')

# Объединение таблиц по ID с добавлением суффиксов для избегания конфликтов имен столбцов
merged_df = pd.merge(df_artists, df_albums, left_on='ID', right_on='ID Исполнителя', suffixes=('_artist', '_album'))
merged_df = pd.merge(merged_df, df_tracks, left_on='ID_album', right_on='ID альбома', suffixes=('', '_track'))
merged_df = pd.merge(merged_df, df_genres, left_on='ID жанра', right_on='ID', suffixes=('', '_genre'))

# Вычисление коэффициента корреляции Пирсона
correlation, p_value = pearsonr(merged_df['Длительность'], merged_df['Стоимость'])

print(f"Коэффициент корреляции Пирсона: {correlation:.2f}")
print(f"P-значение: {p_value:.2e}")
```

# **Интерпретация результатов:**
1. **Коэффициент корреляции Пирсона: 0.01**
Коэффициент корреляции Пирсона (r) измеряет линейную зависимость между двумя переменными. Значение r может варьироваться от -1 до 1.
r = 1 означает идеальную положительную линейную зависимость.
r = -1 означает идеальную отрицательную линейную зависимость.
r = 0 означает отсутствие линейной зависимости.
Значение 0.01 очень близко к нулю, что указывает на практически полное отсутствие линейной зависимости между длительностью песен и их стоимостью.

2. **P-значение: 0.391**
P-значение (p-value) используется для проверки гипотезы о значимости корреляции.
Обычно, p-значение меньше 0.05 (или 5%) указывает на статистически значимую корреляцию.
P-значение 0.391 значительно выше 0.05, что означает, что наблюдаемая корреляция не является статистически значимой.

# **Вывод:**
* Практически полное отсутствие линейной зависимости: Коэффициент корреляции Пирсона, равный 0.01, указывает на отсутствие линейной зависимости между длительностью песен и их стоимостью. Это значит, что изменения в длительности песен практически не влияют на их стоимость.

* Отсутствие статистической значимости: Высокое p-значение (0.391) указывает на то, что наблюдаемая слабая корреляция не является статистически значимой. Это означает, что нет достаточных доказательств для утверждения, что существует какая-либо зависимость между длительностью песен и их стоимостью.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла Excel
df_artists = pd.read_excel('BD.xlsx', sheet_name='Артисты')
df_albums = pd.read_excel('BD.xlsx', sheet_name='Альбомы')
df_tracks = pd.read_excel('BD.xlsx', sheet_name='Треки')

# Объединение таблиц по ID
merged_df = pd.merge(df_artists, df_albums, left_on='ID', right_on='ID Исполнителя', suffixes=('_artist', '_album'))
merged_df = pd.merge(merged_df, df_tracks, left_on='ID_album', right_on='ID альбома', suffixes=('', '_track'))

# Подсчет общей стоимости альбомов для каждого исполнителя
artist_album_costs = merged_df.groupby('Имя')['Стоимость'].sum()

# Выбор ТОП-5 исполнителей
top_5_artists = artist_album_costs.nlargest(5)

# Вывод текстом
print("ТОП-5 исполнителей по стоимости альбомов:")
print(top_5_artists)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Загрузка данных из Excel файла
file_path = 'BD.xlsx'

# Чтение листов Excel файла
artists_df = pd.read_excel(file_path, sheet_name='Артисты')
albums_df = pd.read_excel(file_path, sheet_name='Альбомы')
tracks_df = pd.read_excel(file_path, sheet_name='Треки')
genres_df = pd.read_excel(file_path, sheet_name='Жанры')

# Объединение данных
tracks_albums_df = pd.merge(tracks_df, albums_df, left_on='ID альбома', right_on='ID', suffixes=('_track', '_album'))
tracks_albums_genres_df = pd.merge(tracks_albums_df, genres_df, left_on='ID жанра', right_on='ID', suffixes=('_track_album', '_genre'))

# Убираем ненужные колонки
tracks_albums_genres_df = tracks_albums_genres_df[['Название трека', 'Длительность', 'Размер', 'Стоимость', 'Название_genre']]

# Переименуем колонки для удобства
tracks_albums_genres_df.rename(columns={'Название_genre': 'Жанр'}, inplace=True)

# Просмотр объединенных данных
tracks_albums_genres_df.head()
```

```python
# Преобразование длительности из миллисекунд в секунды
tracks_albums_genres_df['Длительность'] = tracks_albums_genres_df['Длительность'] / 1000
```

```python
# Построение графиков для каждого жанра
genres = tracks_albums_genres_df['Жанр'].unique()

for genre in genres:
    genre_data = tracks_albums_genres_df[tracks_albums_genres_df['Жанр'] == genre]
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=genre_data, x='Длительность', y='Стоимость')
    plt.title(f'Влияние длительности трека на его стоимость для жанра {genre}')
    plt.xlabel('Длительность (секунды)')
    plt.ylabel('Стоимость')
    plt.grid(True)
    plt.show()
```

```python
# Расчет корреляции
correlation_by_genre = tracks_albums_genres_df.groupby('Жанр').apply(lambda x: x[['Длительность', 'Стоимость']].corr().iloc[0, 1])
print("Коэффициент корреляции Пирсона между длительностью и стоимостью по жанрам:")
print(correlation_by_genre)
```
