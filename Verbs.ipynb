```python
import random

# создаем словарь неправильных глаголов и их перевод на английский
irregular_verbs = {
    'сгибать': ['bend', 'bent', 'bent'],
    'делать ставку': ['bet', 'bet', 'bet'],
    'кусать': ['bite', 'bit', 'bitten'],
    'истекать кровью': ['bleed', 'bled', 'bled'],
    'гореть': ['burn', 'burnt', 'burnt'],
    'стоить': ['cost', 'cost', 'cost'],
    'иметь дело': ['deal', 'dealt', 'dealt'],
    'копать': ['dig', 'dug', 'dug'],
    'простить': ['forgive', 'forgave', 'forgiven'],
    'заморозить': ['freeze', 'froze', 'frozen'],
    'тонуть': ['sink', 'sank', 'sunk'],
    'пахнуть': ['smell', 'smelt', 'smelt'],
    'очаровывать': ['spell', 'spelt', 'spelt'],
    'воровать': ['steal', 'stole', 'stolen'],
    'рвать': ['tear', 'tore', 'torn'],
    'бросать': ['throw', 'threw', 'thrown'],
    'будить': ['wake', 'woke', 'woken'],
    'носить': ['wear', 'wore', 'worn'],
    'звонить': ['ring', 'rang', 'rung'],
    'подниматься': ['rise', 'rose', 'risen'],
    'продавать': ['sell', 'sold', 'sold'],
    'отправлять': ['send', 'sent', 'sent'],
    'светить': ['shine', 'shone', 'shone'],
    'стрелять': ['shoot', 'shot', 'shot'],
    'запирать': ['shut', 'shut', 'shut'],
    'вешать': ['hang', 'hung', 'hung'],
    'скрывать': ['hide', 'hid', 'hidden'],
    'ударять': ['hit', 'hit', 'hit'],
    'держать': ['hold', 'held', 'held'],
    'повреждать': ['hurt', 'hurt', 'hurt'],
    'класть': ['lay', 'laid', 'laid'],
    'приводить': ['lead', 'led', 'led'],
    'одалживать': ['lend', 'lent', 'lent'],
    'лгать': ['lie', 'lay', 'lay'],
    'бить': ['beat', 'beat', 'beaten'],
    # добавьте свои глаголы и их 2 и 3 формы
}

# функция для тренировки глагола
def practice_verb(verb, correct_forms):
    # выводим глагол на русском языке
    print("Глагол:", verb)
    # запрашиваем у пользователя правильные формы
    user_form_1 = input("Введите 1 форму глагола на английском: ")
    user_form_2 = input("Введите 2 форму глагола на английском: ")
    user_form_3 = input("Введите 3 форму глагола на английском: ")
    # проверяем ответы пользователя
    if user_form_1.lower() == correct_forms[0].lower() and user_form_2.lower() == correct_forms[1].lower() and user_form_3.lower() == correct_forms[2].lower():
        print("Правильно!")
    else:
        print("Неправильно. Правильные формы:", correct_forms[0], "и", correct_forms[1], "и", correct_forms[2])

# бесконечный цикл для тренировки глаголов
while True:
    # выбираем случайный глагол из словаря
    verb = random.choice(list(irregular_verbs.keys()))
    # выбираем правильные 2 и 3 формы для этого глагола
    correct_forms = irregular_verbs[verb]
    # тренируем глагол
    practice_verb(verb, correct_forms)
    # запрашиваем у пользователя, хочет ли он продолжить тренировку
    again = input("Хотите продолжить тренировку? (y/n) ")
    # если пользователь отвечает "n", выходим из цикла
    if again.lower() != "y":
        break
```
