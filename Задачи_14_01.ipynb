# Задача 1

```python
# Ввод данных
K = int(input("Введите число K: "))
N = int(input("Введите размер матрицы N: "))

# Загрузка файла в Google Colab (загружаете файл с числами)
from google.colab import files
uploaded = files.upload()

# Чтение матрицы из загруженного файла
file_name = list(uploaded.keys())[0]  # Получаем имя загруженного файла

# Проверка существования файла и его содержимого
try:
    with open(file_name, "r") as file:
        # Считываем все числа из файла в одну строку и разделяем на элементы
        elements = list(map(int, file.read().split()))
except FileNotFoundError:
    print(f"Файл {file_name} не найден.")
    exit()

# Проверяем, хватает ли элементов для формирования матрицы
if len(elements) < N * N:
    print("Недостаточно элементов в файле для формирования матрицы.")
    exit()

# Формирование квадратной матрицы A(N, N)
A = [elements[i * N:(i + 1) * N] for i in range(N)]

# Вывод матрицы
print("Матрица A:")
for row in A:
    print(row)

# Деление матрицы на 4 части
part_1 = []  # Часть 1 (левый нижний треугольник)
part_2 = []  # Часть 2 (верхний треугольник)
part_3 = []  # Часть 3 (правый нижний треугольник)
part_4 = []  # Часть 4 (нижний треугольник)

for i in range(N):
    for j in range(N):
        if i > j and i + j < N - 1:
            part_1.append(A[i][j])
        elif i < j and i + j < N - 1:
            part_2.append(A[i][j])
        elif i < j and i + j > N - 1:
            part_3.append(A[i][j])
        elif i > j and i + j > N - 1:
            part_4.append(A[i][j])

# Вывод частей матрицы
print("\nЧасть 1 (Левый нижний треугольник):", part_1)
print("Часть 2 (Верхний треугольник):", part_2)
print("Часть 3 (Правый нижний треугольник):", part_3)
print("Часть 4 (Нижний треугольник):", part_4)
```

# Задача 2

```python
# Ввод данных
K = int(input("Введите число K: "))
N = int(input("Введите размер матрицы N: "))

# Проверка, что N корректно для областей (должно быть >= 4 и чётное)
if N < 4 or N % 2 != 0:
    print("Размер матрицы N должен быть четным и >= 4.")
    exit()

# Формирование матрицы A случайными числами в диапазоне [-10, 10]
import random
A = [[random.randint(-10, 10) for _ in range(N)] for _ in range(N)]

# Вывод матрицы A
print("Матрица A:")
for row in A:
    print(row)

# Копирование A в F
F = [row.copy() for row in A]

# Считаем нулевые элементы в нечётных столбцах области 4
zeros_in_4 = 0
for i in range(N // 2, N):
    for j in range(N // 2):
        if j % 2 == 0 and F[i][j] == 0:
            zeros_in_4 += 1

# Считаем отрицательные элементы в чётных строках области 2
negatives_in_2 = 0
for i in range(N // 2):
    for j in range(N // 2, N):
        if i % 2 == 1 and F[i][j] < 0:
            negatives_in_2 += 1

# Условие для замены областей
if zeros_in_4 > negatives_in_2:
    # Меняем области 4 и 3 симметрично
    for i in range(N // 2, N):
        for j in range(N // 2):
            F[i][j], F[i][N - 1 - j] = F[i][N - 1 - j], F[i][j]
else:
    # Меняем области 1 и 2 несимметрично
    for i in range(N // 2):
        for j in range(N // 2):
            F[i][j], F[i][j + N // 2] = F[i][j + N // 2], F[i][j]

# Вывод матрицы F после преобразований
print("\nМатрица F:")
for row in F:
    print(row)

# Транспонирование матрицы A
AT = [[A[j][i] for j in range(N)] for i in range(N)]

# Вычисление ((F + A) - (K * F)) * AT
FA_sum = [[F[i][j] + A[i][j] for j in range(N)] for i in range(N)]
KF = [[K * F[i][j] for j in range(N)] for i in range(N)]
FA_diff = [[FA_sum[i][j] - KF[i][j] for j in range(N)] for i in range(N)]

# Результат умножения на AT
result = [[sum(FA_diff[i][k] * AT[k][j] for k in range(N)) for j in range(N)] for i in range(N)]

# Вывод результата
print("\nРезультат ((F + A) - (K * F)) * AT:")
for row in result:
    print(row)
```

# Задача 3

```python
import re

# Загрузим файл через Google Colab
from google.colab import files

uploaded = files.upload()

# Функция для преобразования чисел
def process_number(number):
    # Для вещественных чисел меняем точку на запятую
    return number.replace('.', ',')

# Функция для обработки строки из файла
def process_line(line):
    # Разбиваем строку по пробелам
    objects = line.split()
    result = []

    for obj in objects:
        # Если объект это число (вещественное или целое), обрабатываем его
        if re.match(r'^-?\d+(\.\d+)?$', obj):  # Проверка на числовой формат
            result.append(process_number(obj))
        else:
            result.append(obj)  # Если это не число, добавляем без изменений

    return ' '.join(result)

# Чтение и обработка файла
for filename in uploaded.keys():
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            print(process_line(line))
```

# Задача 4

Вывод может быть очень большим, так как существует 10,000,000 возможных вариантов меню. В случае с таким количеством вариантов вывод лучше ограничить:

```python
import itertools

# Список фруктов
fruits = ['яблоко', 'банан', 'груша', 'апельсин', 'киви', 'ананас', 'вишня', 'персик', 'манго', 'слива']

# Генерируем все возможные варианты меню на 7 дней (по одному фрукту на каждый день)
menu_variants = itertools.product(fruits, repeat=7)

# Ограничим вывод первыми 10 вариантами
for i, menu in enumerate(menu_variants):
    if i >= 10:
        break
    print(menu)
```

Либо исключим ситуации, где один и тот же фрукт повторяется два или более раз за неделю

```python
import itertools

# Список фруктов
fruits = ['яблоко', 'банан', 'груша', 'апельсин', 'киви', 'ананас', 'вишня', 'персик', 'манго', 'слива']

# Генерируем все возможные перестановки из 7 фруктов
menu_variants = itertools.permutations(fruits, 7)

# Ограничим вывод первыми 10 вариантами
for i, menu in enumerate(menu_variants):
    if i >= 10:
        break
    print(menu)
```

# Задача 5

```python
import math
import time
import pandas as pd

# Рекурсивная версия функции
def F_recursive(n, memo={0: 1, 1: 2}):
    if n not in memo:
        memo[n] = (-1)**n * (F_recursive(n - 1, memo) / math.factorial(n) - 2 * F_recursive(n - 2, memo) / math.factorial(2 * n))
    return memo[n]

# Итеративная версия функции
def F_iterative(n):
    F_values = {0: 1, 1: 2}
    for i in range(2, n + 1):
        F_values[i] = (-1)**i * (F_values[i - 1] / math.factorial(i) - 2 * F_values[i - 2] / math.factorial(2 * i))
    return F_values[n]

# Тестирование времени выполнения
n_values = list(range(5, 21))  # Проверим для n от 5 до 20
recursive_times = []
iterative_times = []
for n in n_values:
    # Замер времени для рекурсивного подхода
    start_time = time.time()
    F_recursive(n)
    recursive_times.append(time.time() - start_time)

    # Замер времени для итеративного подхода
    start_time = time.time()
    F_iterative(n)
    iterative_times.append(time.time() - start_time)

# Создание DataFrame для представления результатов
df = pd.DataFrame({
    'n': n_values,
    'Recursive Time (seconds)': recursive_times,
    'Iterative Time (seconds)': iterative_times
})

print(df)
```
